<font color='#364C6F'><h1><b>Classification Overview</b></h1>
<h3>In general words classification is to organize things into groups, as it help us understand the relationships between two members of the same group based on their similarity and also help us identify two groups based on their dis-similarity</h3></font>
<font color='#364C6F'>The classification technique can be applied in Machine Learning and statistics which can be used for assigning input vectors to one of the K discrete classes. The classes have to be mutually exclusive to make sure that each input value belongs to a single class. Classification can be used to  predict a class or ategory the data belongs to. It is widely used for different real-life problems like Spam Detection, Churn Prediction, Sentiment Analysis, Dog Breed Detection, Image Classification.<br>
There are multiple type of Classification techniques.<br>
    Here, as part of this project we will apply KNN which stands for K-Nearest-Neighbours, which works on the principle of identifying K-Nearest Objects for a given object based on their similarity on different parameters</font>

### <font color='#50D2C2'><b>Task 6a Importing the Libraries</b></font>
<div ><font color='#050706'>
<br>
Import the KNeighborsClassifier from the sklearn.neighbors module and accuracy_score from sklearn.metrics module
</font></div>

```python
# Write your code here
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.metrics import accuracy_score 
```

### <font color='#50D2C2'><b>Task 6b - Intialising the Variables</b></font>
<div ><font color='#050706'>
<br>
Create two variables x and y, which hold the independent(petal_width, petal_height, sepal_width and sepal_height) and dependent(class) variables respectively
</font></div>

```python
# Write your code here
column = ['petal_width', 'petal_length', 'sepal_width','sepal_length']
x = idata[['petal_width', 'petal_length', 'sepal_width','sepal_length']]
y = idata['class']
```

### <font color='#50D2C2'><b>Task 6c - Model Building</b></font>
<div ><font color='#050706'>
<br>
Create a variable called classfier which holds the instance of KNeighborsClassifier for nearest number of neighbours as 7. Fit the above model for the variables x and y
</font></div>

```python
# Write your code here
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(x,y)
```

### <font color='#50D2C2'><b>Task 6d - Prediction</b></font>
<div ><font color='#050706'>
Predict the y values for all the x values from the variable x and store the result in variables called y_pred. <br>
</font></div>

```python
# Write your code here
y_pred = classifier.predict(x)
y_pred
```

### <font color='#50D2C2'><b>Task 6e Model Performance</b></font>
<div ><font color='#050706'>
Find the Accuracy of the model using the actual y and predicted y values.<br>
</font><font color='#FF3366'>Interpret the result in your words in the Description cell below
</font></div>

```python
# Write your code here
y_new = []
accu = classifier.predict_proba(x)
for i in range(len(y)):
    if y[i] == 'Iris-setosa':
        y_new.append([1.0,0,0])
    elif y[i] == 'Iris-versicolor':
        y_new.append([0,1.0,0])
    else:
        y_new.append([0,0,1.0])
array = np.array(y_new)
```

Description - to pridict he probability and also to generate numpy array

### <font color='#50D2C2'><b>Task 6f Generalisng the Model</b></font>
<div ><font color='#050706'>
We need to find out the optimal no. of neighbors. For this we need to repeat the steps given in 6a to 6c and check which model gives the highest Accuracy.<br>
Find the optimal no. of neighbors using KNN classifier from number of neighbors  lying 2 to 20 and find out the optimal number of neighbors. </font><font color='#FF3366'>Interpret the result in your words in the Description cell below
    </font></div>

```python
# Write your code here

from math import sqrt
from sklearn.metrics import mean_squared_error 
error_list = []
for i in range(2,21):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(x,y)
    y_pred = classifier.predict(x)
    accu = classifier.predict_proba(x)
    error = sqrt(mean_squared_error(array,accu))
    error_list.append(error)
    print('for value of k ' +str(i) + ' error value = ' + str(error))
i = []
for j in range(2,21):
    i.append(j)
mat.scatter(x = i , y = error_list)
```
